```python
from logquery import allcollection
from jsobj import Object
import numpy
import math

condition_to_order = [
  [ 'interactive', 'link', 'none' ],
  [ 'interactive', 'none', 'link' ],
  [ 'link', 'interactive', 'none' ],
  [ 'link', 'none', 'interactive' ],
  [ 'none', 'interactive', 'link' ],
  [ 'none', 'link', 'interactive' ],
]

def quiznum_taken(vocab, quiztype):
  if quiztype == 'pretest':
    return {
      'japanese1': 0,
      'japanese2': 1,
      'japanese3': 2,
    }[vocab]
  else:
    return {
      'japanese1': 0,
      'japanese2': 1,
      'japanese3': 2,
    }[vocab]

studyformat_to_user_pretest = Object({
  'link': {},
  'interactive': {},
  'none': {},
})
studyformat_to_user_posttest = Object({
  'link': {},
  'interactive': {},
  'none': {},
})

#def user_specific_tweaks(data):
#  if data.username == 'msb@cs.stanford.edu':

def isbanned_user(username):
  if username == None:
    return True
  if username == 'Anonymous User':
    return True
  if username == 'Rachel Cao':
    return True
  if 'geza' in username.lower():
    return True
  return False

user_to_condition = {}

for data in allcollection('quiz'):
  data = Object(data)
  if data.condition != None and data.username != None:
    user_to_condition[data.username] = data.condition

for data in allcollection('quiz'):
  data = Object(data)
  correct = len([x for x in data.answers if x.iscorrect])
  data.correct = correct
  unanswered = len([x for x in data.answers if x.myanswer == '--- select a word ---'])
  data.unanswered = unanswered
  vocab = data.vocab
  if vocab == None:
    vocab = data.lang
  if vocab == 'chinese1':
    continue
  if vocab == None:
    #print 'missing vocab:', data.username
    vocab = 'japanese1'
  condition = data.condition
  if isbanned_user(data.username):
    continue
  condition = int(user_to_condition[data.username])
  #studyformat = data.format # note this is incorrect for pretests: shows the previous format, not the upcoming one
  #studyformat2 = None
  studyformat = condition_to_order[condition][quiznum_taken(vocab, data.quiztype)]
  if studyformat == 'interactive' and data.username == 'Niloufar Salehi':
    continue
  if studyformat == 'link' and data.username == 'Tracey Lin':
    continue
  #if studyformat != studyformat2:
  #  if vocab != 'japanese1' and data.quiztype != 'pretest':
  #    print data.username, vocab, condition, studyformat, studyformat2, data.quiztype, correct, unanswered
  #if studyformat != studyformat2 and data.quiztype != 'pretest':
  #  #print data
  #print data.username, vocab, condition, studyformat, data.quiztype, correct, unanswered
  if data.quiztype == 'pretest':
    studyformat_to_user_pretest[studyformat][data.username] = data
  if data.quiztype == 'posttest':
    #if data.username not in studyformat_to_user_pretest[studyformat]:
    #  print data.username, vocab, condition, studyformat, data.quiztype, correct, unanswered
    studyformat_to_user_posttest[studyformat][data.username] = data

studyformat_to_gains = {
  'interactive': [],
  'link': [],
  'none': [],
}

studyformat_to_pretest = {
  'interactive': [],
  'link': [],
  'none': [],
}

studyformat_to_posttest = {
  'interactive': [],
  'link': [],
  'none': [],
}

studyformat_to_user_gains = {
  'interactive': {},
  'link': {},
  'none': {},
}

from scipy.stats import ttest_ind, ttest_rel

for studyformat,v in studyformat_to_user_posttest.items():
  for username,posttest_data in v.items():
    pretest_data = studyformat_to_user_pretest[studyformat][username]
    #if pretest_data == None:
    #  print username
    #print username, studyformat, 'pretest', pretest_data.correct, pretest_data.unanswered
    #print username, studyformat, 'posttest', posttest_data.correct, posttest_data.unanswered
    pretest = pretest_data.correct
    posttest = posttest_data.correct
    gain = posttest_data.correct - pretest_data.correct
    #print username, studyformat, 'gain', gain
    studyformat_to_user_gains[studyformat][username] = gain
    studyformat_to_gains[studyformat].append(gain)
    studyformat_to_pretest[studyformat].append(pretest)
    studyformat_to_posttest[studyformat].append(posttest)

    
    
def stderr(v):
  return numpy.std(v) / float(math.sqrt(len(v)))

print 'gains:'
for k,v in studyformat_to_gains.items():
  print k, numpy.mean(v), stderr(v)
print 'pretest:'
for k,v in studyformat_to_pretest.items():
  print k, numpy.mean(v), stderr(v)
print 'posttest:'
for k,v in studyformat_to_posttest.items():
  print k, numpy.mean(v), stderr(v)
#print 't-tests:'
#users_to_compare = ['Elaine Zhou', 'Tracey Lin', 'Niloufar Salehi']
#print ttest_rel([studyformat_to_user_gains['interactive'][x] for x in users_to_compare], [studyformat_to_user_gains['link'][x] for x in users_to_compare])
print ttest_ind(studyformat_to_gains['interactive'], studyformat_to_gains['link'])
#print ttest_ind(studyformat_to_posttest['interactive'], studyformat_to_posttest['link'])
#print ttest_rel(studyformat_to_gains['interactive'], studyformat_to_gains['link'])
```
